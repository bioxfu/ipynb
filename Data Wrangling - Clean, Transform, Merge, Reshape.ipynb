# 1. Combining and Merging Data Sets

## 1.1 Database-style DataFrame Merges

```python
from pandas import Series, DataFrame
import pandas as pd
import numpy as np
```

```python
df1 = DataFrame({'key': ['b','b','a','c','a','a','b'], 
                 'data1': range(7)})
```

```python
df2 = DataFrame({'key': ['a','b','d'], 
                 'data2': range(3)})
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2, on='key')
```

```python
df3 = DataFrame({'lkey': ['b','b','a','c','a','a','b'], 
                 'data1': range(7)})
```

```python
df4 = DataFrame({'rkey': ['a','b','d'], 
                 'data2': range(3)})
```

```python
pd.merge(df3, df4, left_on='lkey', right_on='rkey')
```

```python
pd.merge(df1, df2, on='key', how='left')
```

```python
df1 = DataFrame({'key': ['b','b','a','c','a','b'], 
                 'data1': range(6)})
```

```python
df2 = DataFrame({'key': ['a','b','a','b','d'], 
                 'data2': range(5)})
```

```python
pd.merge(df1, df2, on='key', how='inner').sort_index(by='key')
```

```python
left = DataFrame({'key1': ['foo', 'foo', 'bar'],
                  'key2': ['one', 'two', 'one'],
                  'lval': [1, 2, 3]})
```

```python
right = DataFrame({'key1': ['foo', 'foo', 'bar', 'bar'],
                  'key2': ['one', 'one', 'one', 'two'],
                  'rval': [4, 5, 6, 7]})
```

```python
left
```

```python
right
```

```python
pd.merge(left, right, on=['key1', 'key2'], how='outer')
```

```python
pd.merge(left, right, on='key1')
```

```python
pd.merge(left, right, on='key1', suffixes=('_left','_right'))
```

## 1.2 Merging on Index

```python
left1 = DataFrame({'key': ['a','b','a','a','b','c'], 
                 'value': range(6)})
```

```python
right1 = DataFrame({'group_val': [3.5, 7]}, index=['a', 'b'])
```

```python
left1
```

```python
right1
```

```python
pd.merge(left1, right1, left_on='key', right_index=True)
```

```python
pd.merge(left1, right1, left_on='key', right_index=True, how='outer')
```

```python
lefth = DataFrame({'key1': ['Ohio','Ohio','Ohio','Nevada','Nevada'], 
                   'key2': [2000,2001,2002,2001,2002],
                   'data': np.arange(5.)})
```

```python
righth = DataFrame(np.arange(12).reshape(6, 2), 
                   index=[['Nevada','Nevada', 'Ohio','Ohio','Ohio','Ohio'],
                          [2001,2000,2000,2000,2001,2002]], 
                   columns=['event1', 'event2'])
```

```python
lefth
```

```python
righth
```

```python
pd.merge(lefth, righth, left_on=['key1', 'key2'], right_index=True, how='outer')
```

```python
left2 = DataFrame([[1., 2.], [3., 4.], [5., 6.]], index=['a','c','e'],
                   columns=['Ohio', 'Nevada'])
```

```python
right2 = DataFrame([[7., 8.], [9., 10.], [11., 12.], [13, 14]], 
                   index=['b','c','d','e'],
                   columns=['Missouri', 'Alabama'])
```

```python
left2
```

```python
right2
```

```python
pd.merge(left2, right2, left_index=True, right_index=True, how='outer')
```

```python
left2.join(right2)
```

```python
left1
```

```python
right1
```

```python
left1.join(right1, on='key')
```

```python
another = DataFrame([[7., 8.], [9., 10.], [11., 12.], [16, 17]], 
                   index=['a','c','e','f'],
                   columns=['New York', 'Oregon'])
```

```python
left2
```

```python
right2
```

```python
another
```

```python
left2.join([right2, another], how='outer').fillna(0)
```

## 1.3 Concatenating Along an Axis

```python
s1 = Series([0, 1], index=['a', 'b'])
```

```python
s2 = Series([2, 3, 4], index=['c', 'd', 'e'])
```

```python
s3 = Series([5, 6], index=['f', 'g'])
```

```python
pd.concat([s1, s2, s3])
```

```python
pd.concat([s1, s2, s3], axis=1)
```

```python
s4 = pd.concat([s1 * 5, s3])
```

```python
pd.concat([s1, s4], axis=1)
```

```python
pd.concat([s1, s4], axis=1, join='inner')
```

```python
s1
```

```python
s4
```

```python
pd.concat([s1, s4], axis=1, join_axes=([['a','c','b','e']]))
```

```python
result = pd.concat([s1, s1, s3], keys=['one', 'two', 'three'])
```

```python
result
```

```python
result.unstack()
```

```python
pd.concat([s1, s2, s3], axis=1, keys=['one', 'two', 'three'])
```

```python
pd.concat([df1, df2], axis=1, keys=['level1', 'level2'],
          names=['upper', 'lower'])
```
